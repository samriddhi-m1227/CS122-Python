# Plotting with Python

### Overview:
* Plotting and data visualization are fundamental skills in data analysis. Python offers a wide variety of tools and libraries to create everything from simple line charts to complex, interactive dashboards.
* In this brief session on plotting, we will focus mainly on seaborn library.
* Python provides rich support for data visualization:
    - Matplotlib:
      The foundation of Python plotting. It allows you to create basic plots (line, bar, scatter, histograms, etc.), customize axes, labels, colors, markers, and even create animations.
    - Seaborn:
      Built on top of Matplotlib, Seaborn provides a high-level, user-friendly interface for statistical graphics. It comes with attractive default themes, built-in datasets, and easy-to-use functions for complex plots like box plots, violin plots, heatmaps, and facet grids.
    - Other Libraries:
      While today we focus on Matplotlib and Seaborn, it's useful to know there are interactive libraries like Plotly and Bokeh, which allow for more dynamic visualizations suitable for web applications.



#### Matplotlib - brief
* We will have a brief look at matplotlib before moving to seaborn.
* https://matplotlib.org/stable/tutorials/pyplot.html
*

```python
#simple example with matplotlib
#pyplot is  a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
#plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.scatter(x, y, marker='2', linestyle='-', color='b')
#popular marker options o , ^, s=square, or numbers and characters (have different markers).
#linestyle options 'solid' or '-', 'dotted' or ':' , dashed or '--', dashdot or '-.'
#colors options various characters.

#Can add more to the same plot
#plt.plot(y, x, marker='o', linestyle='-', color='r')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Basic Line Plot')
plt.show()
```

#### adding more plots to same figure
* We can use a Figure. Within which, to create subplots in Matplotlib, the plt.subplot() or plt.subplots() functions are used. These functions allow for the arrangement of multiple plots within a single figure.
* The plt.subplot() function takes three integer arguments: The number of rows of subplots, The number of columns of subplots, and The index of the current subplot.
* Subplots are numbered from 1 to rows * columns, starting from the top left and proceeding row-wise.

```python
# example of two plots in one
plt.figure()
plt.subplot(211)
plt.plot(x, y, marker='2', linestyle='-', color='b')
# ax=plt.gca() #get current axes. or subplot.
# ax.text(2, 6, 'Testing TEXT') #adding text
#ax.xlim(0,10) #setting x and y limits for the plot
#ax.ylim(0,20)
plt.subplot(212)
plt.plot(x, x, marker='o', linestyle='-', color='r')
plt.show()
```

## Seaborn
* Seaborn simplifies statistical plotting.
* Everything done of seaborn can be done with matplotlib.
* It is simply a wrapper on top of matplotlib, internally it is calling matplotlib for plotting.
* It also integrates well with dataframes and has some inbuilt datasets that you can use for practice.

https://seaborn.pydata.org/tutorial/introduction.html


```python

import seaborn as sns
import matplotlib.pyplot as plt

print(sns.get_dataset_names())
```

```python

#We will use the tips dataset like last class.

# Load the dataset
tips = sns.load_dataset('tips')

tips
```

```python
#a simple scatter plot

sns.scatterplot(x='total_bill', y='tip', data=tips) #optional size of points send a series of numerical values,
```

Simpler, we did not set many variables, sns helps with all that.

```python
# histogram plots to vsualize grouped data

sns.histplot(tips['tip'])
# optionally can use kde=True, kernel density estimation

#sns.histplot(tips['tip'],kde=True)
```

```python
#barplots

sns.barplot(x='sex', y='tip', data=tips)
```

```python
#boxplots
sns.boxplot(x="day", y="total_bill", data=tips) # add in hue="sex" to get grouped boxplots
```

```python
#jointplots, scatterplot and histogram together

sns.jointplot(x="tip", y="total_bill", data=tips)
#can add regression lines on plot with, kind=reg
#or kind=kde for density estimations.
```

```python
sns.pairplot(tips)
```

```python
sns.heatmap(tips[['tip','total_bill','size']].corr(), annot=True)
```
