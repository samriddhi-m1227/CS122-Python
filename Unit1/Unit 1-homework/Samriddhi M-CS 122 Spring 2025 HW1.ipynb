#Question 2: Text Analysis and Categorization


You are given a paragraph of text that consists of several sentences. Each sentence contains words, and your task is to analyze this text using strings, tuples, and sets. Your goal is to create a report that includes various string manipulations, categorization of words based on their length, and identification of unique words.

Your tasks:

1) Split the given text into individual sentences.

2) For each sentence, split it into words, removing any punctuation marks and converting everything to lowercase.


3) Categorize Words by Length:
Classify each word based on its length:
* Words with length 3 or less: "Short"
* Words with length between 4 and 6 (inclusive): "Medium"
* Words with length greater than 6: "Long"

4) For each sentence, store the words categorized by their length in a tuple (e.g., ("Short", ["cat", "dog"])).

5) Unique Words Analysis:
Identify the unique words (ignoring case) across the entire text. Use a set to store these unique words.

6) Count how many unique words belong to each length category (Short, Medium, Long).

7) Print the following:
* For each sentence, a tuple containing the sentence number and a set of the unique words in that sentence.
* The total number of unique words across the entire text.
* The number of unique words in each length category (Short, Medium, Long).

# Input is provided in skeleton code. Only print the information for the tasks that explicitly start with the word "print".

# Grading Rubric:
String Manipulation: Correctly handle sentence and word extraction, case normalization, and punctuation removal. (10 points)
Tuples: Proper use of tuples to categorize words by length. (10 points)
Sets: Efficient use of sets to store unique words and count them based on their length categories. (10 points)
Control Flow: Properly use if-else and for loops to perform word length classification. (10 points)
Printing results: Accurately generate and print the required report based on the text analysis. No unnecessary  prints. (10 points)

```python
input_text= "Artificial intelligence (AI) is a branch of computer science that focuses on creating machines capable of performing tasks that would normally require human intelligence. These tasks include problem-solving, learning, decision-making, speech recognition, and visual perception. AI systems are designed to simulate human cognitive functions and can learn from experience, improving their performance over time. One of the main goals of AI is to create intelligent agents that can perform tasks autonomously without constant human intervention. AI technologies have been successfully implemented in various fields, such as robotics, healthcare, finance, and transportation. For example, AI-powered medical diagnostic systems can analyze medical images to detect diseases like cancer, while self-driving cars are capable of navigating streets and traffic without human input."


# your code goes here
sentences=[sentence.strip() for sentence in input_text.split('.') if sentence.strip()] #only keep non-empty strings

cat_words=[]
all_unique_words=set()

for sentence_num, sentence in enumerate(sentences,1):
  cleaned_sentence=''.join(char for char in sentence if char.isalnum() or char.isspace()).lower()
  words=cleaned_sentence.split()

  short_words=[word for word in words if len(word)<=3]
  medium_words=[word for word in words if 4<=len(word)<=6]
  long_words=[word for word in words if len(word)>6]


  cat_words.append(("Short", short_words))
  cat_words.append(("Medium", medium_words))
  cat_words.append(("Long", long_words))


  sentence_unique_words=set(words)
  print((sentence_num, sentence_unique_words))

  all_unique_words.update(words)




short_count=len([word for word in all_unique_words if len(word) <= 3])
medium_count=len([word for word in all_unique_words if 4 <= len(word) <= 6])
long_count=len([word for word in all_unique_words if len(word) > 6])




print(f"The Total number of unique words in the entire text: {len(all_unique_words)}")
print(f"Number of unique short words: {short_count}")
print(f"Number of unique medium words: {medium_count}")
print(f"Number of unique long words: {long_count}")



```
