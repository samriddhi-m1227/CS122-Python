# Mutable and Immutable Objects

This is a slight detour to understand the concept of mutable and immutable objects.

Immutable objects cannot be changed after creation.

Mutable objects can be changed after creation.



Observer the following two code snippets

```python
#immutable
a=10
b=a
c=b
print(id(a), a, id(b),b, id(c),c)
a=5
print(id(a), a, id(b),b, id(c),c)
```

```python
#mutable
d=[1,2,3]
e=d
f=e
print(id(d),d, id(e),e, id(f),f)
d[1]='a'
print(id(d),d, id(e),e, id(f),f)
```

so with mutuable objects, you can chnage them in place. the ID stays the same

In python the following are immutable

Numbers: int, float, complex, bool
Strings: str

Tuples: tuple    (Discussed later)

Frozen sets: frozenset (Discussed later)

Bytes: bytes (Discussed later)

If indices are negative numbers, they indicate positions counting from the end of the list, with –1 being the last position in the list, –2 being the second-to-last position, and so forth.

```python
print(x[-1])

print(x[-2])
```

List Slicing: Instead of entering list[index] to extract the item just after index, enter list[index1:index2] to extract all items including index1 and up to (but not including) index2.

What would be the outputs of the following:

```python
x = ["first", "second", "third", "fourth"]

a = x[1:-1]
print(a)

b = x[0:3]
print(b)

c = x[-4:-1:2]
print(c)
```

```python

d = x[-1:2]
print(d)
e = x[:3]
print(e)

f = x[2:]
print(f)
```

You can extract a sublist using the slicing operation [start:stop:step].

start - start position is inclusive

stop - end position is exclusive

step - the size of the step

```python
s = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
sublist = s[2:7:2]
sublist  #observe we dont  need to use print statements
```

```python
#the following code will reverse the list
s[::-1]
rs = s[::-1]
rs
```

# Tuples


Tuples are data structures that are very similar to lists, but they can’t be modified; they can only be created.

- immutable data structures (no modifications)
- you can add and multiple a tuple but that will be a new tuple object not the original one since its immutable (check id)


```python
# This is a three element tuple
x = ('a', 'b', 'c')
x
```

The operation for lists that doesn't modify itself can also be used for tuples.

```python
tuple1=(1,2,3,4,5)
tuple2=('A','B','C')

tuple3=tuple1+tuple2
tuple3
#try a few operations that work for lists
# write in comments if they are also applicable to tuples
# your test code goes here
tuple4=tuple1*2
tuple4
```

Tuples can be easily converted to lists with the list function. And lists can be converted to tuples with the tuple function.

```python
aList = [1, 2, 3, 4]
aTuple = tuple(aList) # convert a list to a tuple
print(aList)
print(aTuple)

tuple2list = list(aTuple) # convert a tuple to a list
print(tuple2list)
```
