#The while loop

```
while condition:
    body
post-loop
```


````python

#Situation: I want to search the dataset for 6, all values in the dataset are unique.

dataset = [1, 8, 5, 2]
target = 6
index = 0
foundat = -1

while index < len(dataset):
    if dataset[index] == target:
      foundat = index
    index += 1
if foundat != -1:
  print(f"Item found at index {foundat}.")
else:
  print(f"{target} is not found in the data set.")

````

# The for loop

In Python, a for loop iterates over the values returned by any iterable objectâ€”that is, any object that can yield a sequence of values. For example, a for loop can iterate over every element in a list, a tuple, a dictionary, or a string. But an iterable object can also be a special function called range or a special type of function called a generator or a generator expression, which can be quite powerful.




else behaves just like in the case of while-else


We can experiment with along with another useful funtion called ```range``` function.


range() has 3 input arguments, start, stop, and step. If only one argument is provided, then it's considered the stop argument. The stop is exclusive, meaning it doesn't include the stop value. step is optional and the default if 1.

````python
# examples of for and range usage


dataset=list(range(1,30,3))
print(dataset)

````

for loops are also popularly used to unpack tuples.

````python
#without unpacking
somelist = [(1, 2), (3, 7), (9, 5)]
result = 0
for t in somelist:
  result = result + (t[0] * t[1])
print(result)

````

````python
#Hereâ€™s the same thing, but cleaner:
somelist = [(1, 2), (3, 7), (9, 5)]
result = 0
for x, y in somelist:  # this style of accessing a elements is called unpacking
  result = result + (x * y)

print(result)
````

````python
#Hereâ€™s the same thing, but on a list:
somelist = [[1, 2], [3, 7], [9, 5]]
result = 0

#Can you try out the appraoch on a list does it work?
#Can you guess what would happen with a set ?

for x,y in somelist:
  result= result+(x*y)

print(result)
````

# Zip, Generator expressions, List and Dictionary Comprehensions


### The `zip` function
Sometimes, itâ€™s useful to combine two or more iterables before looping over them. The `zip` function takes the corresponding elements from one or more iterables and combines them into tuples until it reaches the end of the shortest iterable:

````python
# Try to zip two list of different sizes
x = [5, 6, 8]
y = ['a','b','c']
z = zip(y, x)
# zip returns a zip object
#print(type(z))
print(z)
# convert the zip object to a list
print(list(z))
````

### List and dictionary comprehensions
The pattern of using a `for` loop to iterate through a `list`, modify or select individual elements, and create a new list or dictionary is very common. Such loops often look a lot like the following:

````python
x = [1, 2, 3, 4]
x_squared = []

for num in x:
    x_squared.append(num * num)

print(x_squared)
````

This sort of situation is so common that Python has a special shortcut for such operations, called a *comprehension*. You can think of a `list` or `dictionary` comprehension as a
one-line for loop that creates a new list or dictionary from a sequence.

The pattern of a list comprehension is as follows:

> ```new_list = [expression1 for variable in old_list if expression2]```

and a dictionary comprehension looks like this:

>```new_dict = {expression1:expression2 for variable in list if expression3}```

In both cases, the heart of the expression is similar to the beginning of a `for` loop -- *for variable in list* â€” with some expression using that variable to create a new
key or value and an _optional conditional_ expression using the value of the variable to select whether itâ€™s included in the new list or dictionary.


````python
#  code above using list comprehension
x = [1, 2, 3, 4]

x_squared = [ num**2 for num in x]
print(x_squared)
````

Dictionary comprehensions are similar, but you need to supply both a key and a value. If you want to do something similar to the previous squared example but have the number be
the key and the numberâ€™s square be the value in a dictionary.



````python
# Use dictionary comprehension to create a dictionary
# with the number be the key and the numberâ€™s square be the value.

x = [1, 2, 3, 4]

x_squared = { num:num * num for num in x }
print(x_squared)
````

### Generator expressions

Generator expressions are similar to list comprehensions. A generator expression looks a lot like a list comprehension, except that in place of square brackets, it uses parentheses. The following example is the generator-expression version of the list comprehension already discussed:

````python
x = [1, 2, 3, 4]
# a generator expression uses parentheses
x_squared = (num * num for num in x)

print(x_squared)
for i in x_squared:
    print(i)


````

The advantage of using a generator expression is that the entire list isnâ€™t generated in memory, so arbitrarily large sequences can be generated with little memory overhead.

### Boolean values and expressions
Python has a `Boolean` object type that can be set to either `True` or `False`. Any expression with a Boolean operation returns `True` or `False`.

#### Most Python objects can be used as Booleans
Python generalizes this idea: `0` or `empty values` are False, and any other values are True. In practical terms,
this means the following:
-  The numbers `0`, `0.0`, and `0+0j` are all `False`; any other number is `True`. **Note: `j` is the [complex number in Python](https://note.nkmk.me/en/python-complex/)**
-  The empty string `""` is `False`; any other string is `True`.
-  The empty list `[]` is `False`; any other list is `True`.
-  The empty dictionary `{}` is False; any other dictionary is True.
-  The empty set `set()` is False; any other set is `True`.
-  The special Python value `None` is always `False`.

````python
# Test the boolean value of a few empty and non-empty data structure

if a:=4: # := operator in Python to allow variable assignments within expressions.
         # It is known as the Walrus Operator.
    print(f"okay, {a} is True")
else:
    print(f"oh, {a} is False")

````

### Comparison and Boolean operators

- normal operators for comparing objects: `<`, `<=`, `>`, `>=`, and so forth.
- `==` is the equality test operator
- `!=` is the "not equal to" test.
- `in` and `not in` operators to test membership in sequences (lists, tuples, strings, and dictionaries)
- `is` and `is not` operators to test whether two objects are the same.




The `==` and `!=` operators test to see whether their operands contains the same values (shallow comparison).
`==` and `!=` are used in most situations, as opposed to `is` and `is not` operators, which test to see whether their operands are the same object (deep comparison):

````python
x = [0]
y = [x, 1]
# test to see with x and y[0] are the same object
print(x is y[0])  # x and y[0] are referencing the same object
print(x == y[0])

for i in range(5):
    x = [0] # x has been assigned to a different object
    print(x == y[0])
    print(x is y[0]) # x and y[0] are no longer referencing the same object



````
